```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import cv2
import numpy as np
import albumentations as A
import pandas as pd
import editdistance
import unicodedata
import os
import json
import matplotlib.pyplot as plt
from collections import defaultdict
from heapq import heappush, heappop
from sklearn.model_selection import train_test_split

# Define Nepali character set
nepali_chars = {
    "vowels": ["अ", "आ", "इ", "ई", "उ", "ऊ", "ऋ", "ॠ", "ए", "ऐ", "ओ", "औ", "अं", "अः"],
    "vowel_diacritics": ["", "ा", "ि", "ी", "ु", "ू", "ृ", "ॄ", "े", "ै", "ो", "ौ", "ं", "ः"],
    "consonants": ["क", "ख", "ग", "घ", "ङ", "च", "छ", "ज", "झ", "ञ", "ट", "ठ", "ड", "ढ", "ण",
                   "त", "थ", "द", "ध", "न", "प", "फ", "ब", "भ", "म", "य", "र", "ल", "व", "श",
                   "ष", "स", "ह"],
    "ligatures": ["क्ष", "त्र", "ज्ञ", "श्र"],
    "nukta_consonants": ["क़", "ख़", "ग़", "ज़", "ड़", "ढ़", "फ़", "य़"],
    "special_symbols": ["ँ", "ं", "ः", "्", "।", "॥", "॰", "ॐ"]
}

all_nepali_chars = set(
    nepali_chars["vowels"] +
    [c for c in nepali_chars["vowel_diacritics"] if c] +
    nepali_chars["consonants"] +
    nepali_chars["ligatures"] +
    nepali_chars["nukta_consonants"] +
    nepali_chars["special_symbols"]
)

# Load dataset
try:
    df = pd.read_csv("/content/extracted_folder/train_data/train_labels.txt", sep='\t', names=["img_path", "label"])
    img_paths = ["/content/extracted_folder/train_data/train_images/" + path for path in df["img_path"].tolist()]
    labels = [unicodedata.normalize('NFC', label) for label in df["label"].tolist()]
    print(f"Loaded {len(img_paths)} images from dataset")
except Exception as e:
    print(f"Error loading dataset: {e}")
    print("Ensure /content/extracted_folder/train_data/train_labels.txt exists with format 'img_path\\tlabel'")
    exit()

# Create character set
def create_char_set(labels):
    chars = set()
    for label in labels:
        # Add each character from the label to the set
        chars.update(label)

    # Ensure ligatures are treated as single characters
    ligatures = nepali_chars["ligatures"]
    for ligature in ligatures:
        if any(ligature in label for label in labels):
            chars.add(ligature)

    chars = sorted(list(chars))
    chars.insert(0, "<blank>")  # Add blank token for CTC loss
    return chars

chars = create_char_set(labels)
char_to_idx = {char: idx for idx, char in enumerate(chars)}
idx_to_char = {idx: char for idx, char in enumerate(chars)}

# Check character coverage
missing_chars = all_nepali_chars - set(chars[1:])
print("Missing characters:", missing_chars)
if missing_chars:
    print("Warning: Add data for missing characters:", missing_chars)

# Dataset class
class NepaliDataset(Dataset):
    def __init__(self, img_paths, labels, char_to_idx, target_height=64, max_width=256):
        self.img_paths = img_paths
        self.labels = labels
        self.char_to_idx = char_to_idx
        self.target_height = target_height
        self.max_width = max_width

    def __len__(self):
        return len(self.img_paths)

    def __getitem__(self, idx):
        # Read the image in grayscale
        img = cv2.imread(self.img_paths[idx], cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"Failed to load: {self.img_paths[idx]}")
            img = np.zeros((self.target_height, self.max_width), dtype=np.uint8)

        # Resize the image while maintaining aspect ratio
        h, w = img.shape
        target_width = int(w * self.target_height / h) if h > 0 else self.max_width
        img = cv2.resize(img, (min(target_width, self.max_width), self.target_height))

        # Normalize the image to [0, 1] range
        img = img / 255.0

        # Add channel dimension (1 for grayscale)
        img = np.expand_dims(img, axis=0)

        # Pad the image to ensure uniform size
        padded_img = np.zeros((1, self.target_height, self.max_width), dtype=np.float32)
        padded_img[:, :, :min(target_width, self.max_width)] = img[:, :, :min(target_width, self.max_width)]

        # Encode the label
        label = self.labels[idx]
        encoded_label = [self.char_to_idx[char] for char in label]
        label_length = len(encoded_label)

        return (
            torch.FloatTensor(padded_img),
            torch.LongTensor(encoded_label),
            torch.LongTensor([label_length])
        )
# Custom collate function for variable-length labels
def ocr_collate_fn(batch):
    images, labels, label_lengths = zip(*batch)
    images = torch.stack(images, 0)
    label_lengths = torch.stack(label_lengths, 0)
    return images, labels, label_lengths

# Function to verify dataset (labels and preprocessed images)
def verify_dataset(dataset, idx_to_char, num_samples=5):
    print("Verifying dataset samples:")
    for i in range(min(num_samples, len(dataset))):
        img, encoded_label, label_length = dataset[i]
        label = dataset.labels[i]

        # Break label into individual characters while respecting ligatures
        characterized_label = []
        j = 0
        while j < len(label):
            # Check if the current position matches a ligature
            matched_ligature = None
            for ligature in nepali_chars["ligatures"]:
                if label.startswith(ligature, j):
                    matched_ligature = ligature
                    break
            if matched_ligature:
                characterized_label.append(matched_ligature)
                j += len(matched_ligature)
            else:
                characterized_label.append(label[j])
                j += 1

        decoded_label = ''.join([idx_to_char[idx.item()] for idx in encoded_label[:label_length.item()]])
        print(f"Sample {i+1}:")
        print(f"  Label: {label} --> {characterized_label}")
        print(f"  Decoded label: {decoded_label}")
        print(f"  Image shape: {img.shape}")

        # Visualize the preprocessed image
        plt.imshow(img.squeeze(), cmap='gray')
        plt.title(f"Label: {label}")
        plt.axis('off')
        plt.show()
# Beam search decoder
def ctc_beam_search_decoder(probs, idx_to_char, beam_width=10, blank_idx=0):
    seq_len, vocab_size = probs.shape
    beams = [(("", 0), 1.0)]  # (prefix, last_char_idx), prob
    for t in range(seq_len):
        next_beams = defaultdict(float)
        for (prefix, last_idx), prob in beams:
            for idx in range(vocab_size):
                new_prob = prob * probs[t, idx]
                if idx == blank_idx or (idx == last_idx and prefix and idx != 0):
                    next_beams[(prefix, last_idx)] += new_prob
                else:
                    new_prefix = prefix + idx_to_char[idx] if idx != 0 else prefix
                    next_beams[(new_prefix, idx)] += new_prob
        beams = sorted(next_beams.items(), key=lambda x: x[1], reverse=True)[:beam_width]
    best_prefix = beams[0][0][0]
    return best_prefix

# CRNN model with attention
class Attention(nn.Module):
    def __init__(self, hidden_size):
        super(Attention, self).__init__()
        self.attn = nn.Linear(hidden_size * 2, hidden_size)
        self.v = nn.Parameter(torch.rand(hidden_size))

    def forward(self, hidden, encoder_outputs):
        energy = torch.tanh(self.attn(encoder_outputs))
        attention = F.softmax(torch.matmul(energy, self.v), dim=1)
        context = attention.unsqueeze(2) * encoder_outputs
        return context.sum(dim=1)

class CRNN(nn.Module):
    def __init__(self, num_chars, rnn_hidden_size=512, dropout=0.3):
        super(CRNN, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv2d(1, 64, 3, stride=1, padding=1), nn.ReLU(),
            nn.MaxPool2d(2, 2),
            self._make_res_block(64, 128),
            nn.MaxPool2d(2, 2),
            self._make_res_block(128, 256),
            nn.MaxPool2d((2, 1), (2, 1)),
            self._make_res_block(256, 512),
            nn.MaxPool2d((2, 1), (2, 1)),
            self._make_res_block(512, 512),
            nn.MaxPool2d((4, 1), (4, 1)),
            nn.Dropout(dropout)
        )
        self.rnn = nn.GRU(512, rnn_hidden_size, num_layers=2, bidirectional=True, batch_first=True, dropout=dropout)
        self.attention = Attention(rnn_hidden_size)
        self.linear = nn.Linear(rnn_hidden_size * 2, num_chars + 1)

    def _make_res_block(self, in_channels, out_channels):
        return nn.Sequential(
            nn.Conv2d(in_channels, out_channels, 3, padding=1), nn.BatchNorm2d(out_channels), nn.ReLU(),
            nn.Conv2d(out_channels, out_channels, 3, padding=1), nn.BatchNorm2d(out_channels),
            nn.Conv2d(out_channels, out_channels, 1) if in_channels != out_channels else nn.Identity()
        )

    def forward(self, x):
        conv = self.cnn(x)
        batch, channels, height, width = conv.size()
        assert height == 1, "CNN output height must be 1"
        conv = conv.squeeze(2)  # remove height dim -> [batch, channels, width]
        conv = conv.permute(0, 2, 1)  # -> [batch, width, channels]
        output, _ = self.rnn(conv)  # output: [batch, width, 2*rnn_hidden_size]
        output = self.linear(output)  # -> [batch, width, num_classes]
        return output

# Prepare data
train_img_paths, val_img_paths, train_labels, val_labels = train_test_split(img_paths, labels, test_size=0.2, random_state=42)
train_dataset = NepaliDataset(train_img_paths, train_labels, char_to_idx)
val_dataset = NepaliDataset(val_img_paths, val_labels, char_to_idx)

# Verify training dataset before training
verify_dataset(train_dataset, idx_to_char, num_samples=5)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, collate_fn=ocr_collate_fn)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False, collate_fn=ocr_collate_fn)

# Training and validation
def train_one_epoch(model, loader, criterion, optimizer):
    model.train()
    total_loss = 0
    for images, labels, label_lengths in loader:
        images = images.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        outputs = outputs.log_softmax(2)
        input_lengths = torch.full((images.size(0),), outputs.size(1), dtype=torch.long)
        targets = torch.cat(labels).to(device)
        target_lengths = label_lengths.squeeze(1).to(device)
        loss = criterion(outputs.permute(1, 0, 2), targets, input_lengths, target_lengths)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(loader)

def validate(model, loader, criterion):
    model.eval()
    total_loss = 0
    with torch.no_grad():
        for images, labels, label_lengths in loader:
            images = images.to(device)
            outputs = model(images)
            outputs = outputs.log_softmax(2)
            input_lengths = torch.full((images.size(0),), outputs.size(1), dtype=torch.long)
            targets = torch.cat(labels).to(device)
            target_lengths = label_lengths.squeeze(1).to(device)
            loss = criterion(outputs.permute(1, 0, 2), targets, input_lengths, target_lengths)
            total_loss += loss.item()
    return total_loss / len(loader)

# Prediction function remains unchanged
def predict_word(model, img_path, char_to_idx, idx_to_char, target_height=64, max_width=256):
    model.eval()
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print(f"Failed to load: {img_path}")
        return ""
    h, w = img.shape
    target_width = int(w * target_height / h) if h > 0 else max_width
    img = cv2.resize(img, (target_width, target_height))
    img = img / 255.0
    img = np.expand_dims(img, axis=0)
    padded_img = np.zeros((1, target_height, max_width), dtype=np.float32)
    padded_img[:, :, :min(target_width, max_width)] = img[:, :, :min(target_width, max_width)]
    img = torch.FloatTensor(padded_img).unsqueeze(0).to(device)
    with torch.no_grad():
        output = model(img).softmax(2)
        output = output.squeeze(0).cpu().numpy()
    predicted_text = ctc_beam_search_decoder(output, idx_to_char, beam_width=10, blank_idx=0)
    return predicted_text
# CER
def cer(pred, target):
    return editdistance.eval(pred, target) / len(target)

# Training loop
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CRNN(num_chars=len(chars), rnn_hidden_size=512).to(device)
criterion = nn.CTCLoss(blank=0, zero_infinity=True)
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=5, factor=0.5)

best_val_loss = float('inf')
patience = 10
counter = 0
for epoch in range(50):
    train_loss = train_one_epoch(model, train_loader, criterion, optimizer)
    val_loss = validate(model, val_loader, criterion)
    print(f"Epoch {epoch+1}, Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")
    scheduler.step(val_loss)
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        try:
            torch.save(model.state_dict(), "/content/nepali_crnn2.pth")
            print("Model saved to /content/nepali_crnn2.pth")
        except Exception as e:
            print(f"Error saving model: {e}")
        counter = 0
    else:
        counter += 1
    if counter >= patience:
        print("Early stopping")
        break

# Save character mappings
try:
    with open("/content/char_to_idx.json", "w", encoding="utf-8") as f:
        json.dump(char_to_idx, f, ensure_ascii=False)
    with open("/content/idx_to_char.json", "w", encoding="utf-8") as f:
        json.dump(idx_to_char, f, ensure_ascii=False)
    print("Character mappings saved")
except Exception as e:
    print(f"Error saving character mappings: {e}")

# Evaluate a couple samples
test_words = [
    (val_img_paths[0], val_labels[0]),
    (val_img_paths[1] if len(val_img_paths) > 1 else val_img_paths[0],
     val_labels[1] if len(val_labels) > 1 else val_labels[0]),
]

for img_path, target in test_words:
    pred = predict_word(model, img_path, char_to_idx, idx_to_char)
    print(f"Target: {target}, Predicted: {pred}, CER: {cer(pred, target):.4f}")

# Load saved model and test
print("\nTesting saved model...")
try:
    model.load_state_dict(torch.load("/content/nepali_crnn2.pth"))
    model.eval()
    test_img_path = test_words[0][0]
    pred = predict_word(model, test_img_path, char_to_idx, idx_to_char)
    print(f"Saved model prediction: {pred}")
except Exception as e:
    print(f"Error loading saved model: {e}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import cv2
import numpy as np
import pandas as pd
import editdistance
import unicodedata
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from collections import defaultdict

# Define Nepali character set
nepali_chars = {
    "vowels": ["अ", "आ", "इ", "ई", "उ", "ऊ", "ऋ", "ॠ", "ए", "ऐ", "ओ", "औ", "अं", "अः"],
    "vowel_diacritics": ["", "ा", "ि", "ी", "ु", "ू", "ृ", "ॄ", "े", "ै", "ो", "ौ", "ं", "ः"],
    "consonants": ["क", "ख", "ग", "घ", "ङ", "च", "छ", "ज", "झ", "ञ", "ट", "ठ", "ड", "ढ", "ण",
                   "त", "थ", "द", "ध", "न", "प", "फ", "ब", "भ", "म", "य", "र", "ल", "व", "श",
                   "ष", "स", "ह"],
    "ligatures": ["क्ष", "त्र", "ज्ञ", "श्र"],
    "nukta_consonants": ["क़", "ख़", "ग़", "ज़", "ड़", "ढ़", "फ़", "य़"],
    "special_symbols": ["ँ", "ं", "ः", "्", "।", "॥", "॰", "ॐ"]
}

all_nepali_chars = set(
    nepali_chars["vowels"] +
    [c for c in nepali_chars["vowel_diacritics"] if c] +
    nepali_chars["consonants"] +
    nepali_chars["ligatures"] +
    nepali_chars["nukta_consonants"] +
    nepali_chars["special_symbols"]
)

# Load dataset
try:
    df = pd.read_csv("/content/extracted_folder/train_data/train_labels.txt", sep='\t', names=["img_path", "label"])
    img_paths = ["/content/extracted_folder/train_data/train_images/" + path for path in df["img_path"].tolist()]
    labels = [unicodedata.normalize('NFC', label) for label in df["label"].tolist()]
    print(f"Loaded {len(img_paths)} images from dataset")
except Exception as e:
    print(f"Error loading dataset: {e}")
    print("Ensure /content/extracted_folder/train_data/train_labels.txt exists with format 'img_path\\tlabel'")
    exit()

# Create character set
def create_char_set(labels):
    chars = set()
    for label in labels:
        chars.update(label)
    ligatures = nepali_chars["ligatures"]
    for ligature in ligatures:
        if any(ligature in label for label in labels):
            chars.add(ligature)
    chars = sorted(list(chars))
    chars.insert(0, "<blank>")  # Add blank token for CTC
    return chars

chars = create_char_set(labels)
char_to_idx = {char: idx for idx, char in enumerate(chars)}
idx_to_char = {idx: char for idx, char in enumerate(chars)}

# Dataset class
class NepaliDataset(Dataset):
    def __init__(self, img_paths, labels, char_to_idx, target_height=64, max_width=256):
        self.img_paths = img_paths
        self.labels = labels
        self.char_to_idx = char_to_idx
        self.target_height = target_height
        self.max_width = max_width

    def __len__(self):
        return len(self.img_paths)

    def __getitem__(self, idx):
        img = cv2.imread(self.img_paths[idx], cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"Failed to load: {self.img_paths[idx]}")
            img = np.zeros((self.target_height, self.max_width), dtype=np.uint8)
        h, w = img.shape
        target_width = int(w * self.target_height / h) if h > 0 else self.max_width
        img = cv2.resize(img, (min(target_width, self.max_width), self.target_height))
        img = img / 255.0
        img = np.expand_dims(img, axis=0)
        padded_img = np.zeros((1, self.target_height, self.max_width), dtype=np.float32)
        padded_img[:, :, :min(target_width, self.max_width)] = img[:, :, :min(target_width, self.max_width)]
        label = self.labels[idx]
        encoded_label = [self.char_to_idx[char] for char in label]
        label_length = len(encoded_label)
        return (
            torch.FloatTensor(padded_img),
            torch.LongTensor(encoded_label),
            torch.LongTensor([label_length])
        )

# Custom collate function
def ocr_collate_fn(batch):
    images, labels, label_lengths = zip(*batch)
    images = torch.stack(images, 0)
    label_lengths = torch.stack(label_lengths, 0)
    return images, labels, label_lengths

# Attention module
class Attention(nn.Module):
    def __init__(self, hidden_size):
        super(Attention, self).__init__()
        self.attn = nn.Linear(hidden_size * 2, hidden_size)
        self.v = nn.Parameter(torch.rand(hidden_size))

    def forward(self, hidden, encoder_outputs):
        energy = torch.tanh(self.attn(encoder_outputs))
        attention = F.softmax(torch.matmul(energy, self.v), dim=1)
        context = attention.unsqueeze(2) * encoder_outputs
        return context.sum(dim=1)

# CRNN model with attention
class CRNN(nn.Module):
    def __init__(self, num_chars, rnn_hidden_size=512, dropout=0.3):
        super(CRNN, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv2d(1, 64, 3, stride=1, padding=1), nn.ReLU(),
            nn.MaxPool2d(2, 2),
            self._make_res_block(64, 128),
            nn.MaxPool2d(2, 2),
            self._make_res_block(128, 256),
            nn.MaxPool2d((2, 1), (2, 1)),
            self._make_res_block(256, 512),
            nn.MaxPool2d((2, 1), (2, 1)),
            self._make_res_block(512, 512),
            nn.MaxPool2d((4, 1), (4, 1)),
            nn.Dropout(dropout)
        )
        self.rnn = nn.GRU(512, rnn_hidden_size, num_layers=2, bidirectional=True, batch_first=True, dropout=dropout)
        self.attention = Attention(rnn_hidden_size)
        self.linear = nn.Linear(rnn_hidden_size * 2, num_chars + 1)

    def _make_res_block(self, in_channels, out_channels):
        return nn.Sequential(
            nn.Conv2d(in_channels, out_channels, 3, padding=1), nn.BatchNorm2d(out_channels), nn.ReLU(),
            nn.Conv2d(out_channels, out_channels, 3, padding=1), nn.BatchNorm2d(out_channels),
            nn.Conv2d(out_channels, out_channels, 1) if in_channels != out_channels else nn.Identity()
        )

    def forward(self, x):
        conv = self.cnn(x)
        batch, channels, height, width = conv.size()
        assert height == 1, "CNN output height must be 1"
        conv = conv.squeeze(2)
        conv = conv.permute(0, 2, 1)
        output, _ = self.rnn(conv)
        output = self.linear(output)
        return output

# Beam search decoder
def ctc_beam_search_decoder(probs, idx_to_char, beam_width=10, blank_idx=0):
    seq_len, vocab_size = probs.shape
    beams = [(("", 0), 1.0)]  # (prefix, last_char_idx), prob
    for t in range(seq_len):
        next_beams = defaultdict(float)
        for (prefix, last_idx), prob in beams:
            for idx in range(vocab_size):
                if idx not in idx_to_char:  # Skip invalid indices
                    continue
                new_prob = prob * probs[t, idx]
                if idx == blank_idx or (idx == last_idx and prefix and idx != 0):
                    next_beams[(prefix, last_idx)] += new_prob
                else:
                    new_prefix = prefix + idx_to_char[idx] if idx != 0 else prefix
                    next_beams[(new_prefix, idx)] += new_prob
        beams = sorted(next_beams.items(), key=lambda x: x[1], reverse=True)[:beam_width]
    best_prefix = beams[0][0][0] if beams else ""
    return best_prefix

# CER and WER
def cer(pred, target):
    return editdistance.eval(pred, target) / len(target) if len(target) > 0 else 0.0

def wer(pred, target):
    pred_words = pred.split()
    target_words = target.split()
    if len(target_words) == 0:
        return 1.0 if len(pred_words) > 0 else 0.0
    return editdistance.eval(pred_words, target_words) / len(target_words)

# Evaluation function
def evaluate_model(model, dataset, char_to_idx, idx_to_char, device, batch_size=16):
    model.eval()
    loader = DataLoader(dataset, batch_size=batch_size, shuffle=False, collate_fn=ocr_collate_fn)

    cer_scores = []
    wer_scores = []
    predictions = []
    ground_truths = []

    with torch.no_grad():
        for batch_idx, (images, _, _) in enumerate(tqdm(loader, desc="Evaluating")):
            images = images.to(device)
            batch_size = images.size(0)
            outputs = model(images).softmax(2)
            outputs = outputs.cpu().numpy()

            for i in range(batch_size):
                output = outputs[i]
                pred_text = ctc_beam_search_decoder(output, idx_to_char, beam_width=10, blank_idx=0)
                gt_text = dataset.labels[batch_idx * batch_size + i]
                cer_score = cer(pred_text, gt_text)
                wer_score = wer(pred_text, gt_text)
                cer_scores.append(cer_score)
                wer_scores.append(wer_score)
                predictions.append(pred_text)
                ground_truths.append(gt_text)

    avg_cer = np.mean(cer_scores) if cer_scores else 0.0
    avg_wer = np.mean(wer_scores) if wer_scores else 0.0

    return avg_cer, avg_wer, predictions, ground_truths

# Prepare validation dataset
train_img_paths, val_img_paths, train_labels, val_labels = train_test_split(img_paths, labels, test_size=0.2, random_state=42)
val_dataset = NepaliDataset(val_img_paths, val_labels, char_to_idx)

# Load and evaluate model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CRNN(num_chars=len(chars), rnn_hidden_size=512).to(device)
try:
    model.load_state_dict(torch.load("/content/nepali_crnn2.pth", map_location=device))
    print("Model loaded successfully from /content/nepali_crnn2.pth")
except Exception as e:
    print(f"Error loading model: {e}")
    exit()

# Run evaluation
avg_cer, avg_wer, predictions, ground_truths = evaluate_model(model, val_dataset, char_to_idx, idx_to_char, device)

# Print results
print(f"\nEvaluation Results on Validation Set ({len(val_dataset)} samples):")
print(f"Average CER: {avg_cer:.4f}")
print(f"Average WER: {avg_wer:.4f}")

# Print all predictions
print("\nAll Predictions:")
for i, (gt, pred) in enumerate(zip(ground_truths, predictions)):
    print(f"Sample {i+1}:")
    print(f"Ground Truth: {gt}")
    print(f"Predicted: {pred}")
    print(f"CER: {cer(pred, gt):.4f}")
    print("-" * 50)
```
